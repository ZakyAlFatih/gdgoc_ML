```python
import os
import random
import shutil
import pandas as pd

def create_val_folder(train_folder, val_folder, csv_file, val_percentage):
    # Membaca CSV untuk mendapatkan label gambar
    df = pd.read_csv(csv_file)

    # Bersihkan spasi tambahan di nama kolom
    df.columns = df.columns.str.strip()

    # Pastikan kolom yang benar digunakan
    expected_columns = ["image", "label"]  # Sesuai dengan yang terdeteksi di CSV
    for col in expected_columns:
        if col not in df.columns:
            raise KeyError(f"Kolom '{col}' tidak ditemukan dalam CSV. Kolom tersedia: {df.columns.tolist()}")

    # Konversi kolom 'image' menjadi string untuk menghindari error TypeError
    df["image"] = df["image"].astype(str)

    # Membuat folder val jika belum ada
    os.makedirs(val_folder, exist_ok=True)

    # Kelompokkan gambar berdasarkan label
    label_dict = {}
    for _, row in df.iterrows():
        image_name = row["image"]  # Menggunakan 'image' sesuai CSV
        label = str(row["label"])  # Konversi label menjadi string untuk nama folder

        if label not in label_dict:
            label_dict[label] = []
        label_dict[label].append(image_name)

    for label, images in label_dict.items():
        # Membuat subfolder berdasarkan label jika belum ada
        train_label_folder = os.path.join(train_folder, label)
        val_label_folder = os.path.join(val_folder, label)

        os.makedirs(train_label_folder, exist_ok=True)
        os.makedirs(val_label_folder, exist_ok=True)

        # Memindahkan gambar dari train root ke subfolder label
        for image in images:
            src_path = os.path.join(train_folder, image)
            dst_path = os.path.join(train_label_folder, image)

            if os.path.exists(src_path):  # Pastikan file ada sebelum dipindahkan
                shutil.move(src_path, dst_path)

        # Pisahkan gambar untuk validasi
        num_val_images = int(len(images) * val_percentage)
        val_images = random.sample(images, num_val_images)

        for val_image in val_images:
            train_path = os.path.join(train_label_folder, val_image)
            val_path = os.path.join(val_label_folder, val_image)

            if os.path.exists(train_path):
                shutil.move(train_path, val_path)

if __name__ == "__main__":
    train_folder_path = "/content/train"
    val_folder_path = "/content/val"
    train_csv_path = "/content/train.csv"
    val_percentage = 0.1  # 10% gambar untuk validasi

    create_val_folder(train_folder_path, val_folder_path, train_csv_path, val_percentage)
```

```python
import pandas as pd
import os
from ultralytics import YOLO

model = YOLO('yolov8m-cls.pt')
model.train(data='/content', epochs=5)
# Path ke model yang telah dilatih
MODEL_PATH = './runs/classify/train/weights/best.pt'

# Path ke folder test
TEST_FOLDER_PATH = '/content/test'

# Path untuk menyimpan file submission
SUBMISSION_FILE_PATH = '/content/submission.csv'

# Load model YOLO
model = YOLO(MODEL_PATH)

def predict_image(file_path):
    """Fungsi untuk melakukan prediksi pada gambar tunggal."""
    result = model(file_path)
    return result[0].names[result[0].probs.top1]

# Load file test.csv
test_csv_path = '/content/test.csv'
test_df = pd.read_csv(test_csv_path)

# Lakukan prediksi untuk setiap gambar dalam test.csv
predictions = []
for image_name in test_df['image']:
    image_path = os.path.join(TEST_FOLDER_PATH, image_name)
    if os.path.exists(image_path):
        prediction = predict_image(image_path)
        label = 1 if prediction == 'AI-generated' else 0  # Menyesuaikan label dengan format yang diminta
    else:
        label = 0  # Jika file tidak ditemukan, default ke 0
    predictions.append([image_name, label])

# Simpan hasil prediksi dalam format submission
submission_df = pd.DataFrame(predictions, columns=['image', 'label'])
submission_df.to_csv(SUBMISSION_FILE_PATH, index=False)

print(f"Submission file saved at {SUBMISSION_FILE_PATH}")
```
